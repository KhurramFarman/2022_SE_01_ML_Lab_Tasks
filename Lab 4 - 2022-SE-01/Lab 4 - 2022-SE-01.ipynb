```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Dataset .csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

# 1. Bar Chart: Top 10 Cities with the Most Restaurants

```python
plt.figure(figsize=(12, 6))
top_cities = df["City"].value_counts().nlargest(20)
sns.barplot(x=top_cities.index, y=top_cities.values, palette="viridis")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities with the Most Restaurants")
plt.xticks(rotation=45)
plt.show()
```

# Analysis:
- New delhi has the most number of resturants
- its also because that the population of new delhi is 33 million

# 2. Pie Chart: Cuisine Distribution (Top 10 Cuisines)

```python
plt.figure(figsize=(14, 8))
top_cuisines = df["Cuisines"].value_counts().nlargest(15)
plt.pie(top_cuisines, labels=top_cuisines.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Top 10 Cuisines Distribution")
plt.show()
```

# Analysis:
- Indian cusines beats other categories by margins especially North Indian cusines   

# 3. Satter Plot: Restaurant Locations with Rating Encoding

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', data=df)
plt.title('Restaurant Locations with Rating Encoding')
plt.show()
```

# Analysis:
- North America, Austrilia and South East Asia has more and high rated restaurants 

# 4. Box Plot: Price Range vs. Aggregate Rating

```python
plt.figure(figsize=(7,5))
sns.boxplot(x=df["Price range"], y=df["Aggregate rating"], palette="coolwarm")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.title("Price Range vs. Aggregate Rating")
plt.show()
```

# Analysis
- Expensive restaurants usually get better and more consistent ratings. Cheap restaurants have mixed ratings, with some being very good and others very bad.

# 5. Scatter Plot: Aggregate Rating vs. Number of Votes

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x=df["Aggregate rating"], y=df["Votes"], alpha=0.5, color="green")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Votes")
plt.title("Aggregate Rating vs. Number of Votes")
plt.show()
```

# Analysis
- Restaurants with higher ratings tend to receive more votes. Lower-rated restaurants usually have fewer votes, indicating less customer engagement.

# 6. Stacked Bar Chart: Table Availability vs. Online Booking

```python
table_online = df.groupby(['Has Table booking', 'Has Online delivery']).size().unstack()
table_online.plot(kind="bar", stacked=True, figsize=(6, 4))
plt.title("Table Reservation vs. Online Booking")
plt.xlabel("Has Table")
plt.ylabel("Count")
plt.show()
```

# Analysis:
- Most restaurants do not offer table reservations, and among them, a significant portion provides online delivery. Restaurants with table reservations are fewer and have a lower share of online delivery.

# 7. Count Plot: Rating Categories

```python
plt.figure(figsize=(7, 5))
sns.countplot(x=df['Rating text'])
plt.title("Count of Rating Categories")
plt.xlabel("Rating Category")
plt.ylabel("Count")
plt.show()
```

# Analysis:
- Most restaurants fall under the "Average" rating category, followed by "Not rated" and "Good." Very few restaurants are rated as "Excellent" or "Poor."

# 8. Box Plot: Has Table Booking vs. Ratings

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Has Table booking'], y=df['Aggregate rating'])
plt.xlabel('Has Table Booking')
plt.ylabel('Ratings')
plt.title('Table Booking vs Ratings')
plt.show()
```

#  Analysis:
- Restaurants with table booking generally have higher ratings, with fewer low-rated outliers.
- Restaurants without table booking show a wider spread, including many low-rated ones.

# 9. Bar Plot Top 10 Most Rated Restaurants

```python
plt.figure(figsize=(10,5))
top_rated = df[['Restaurant Name', 'Votes']].sort_values(by='Votes', ascending=False).head(10)
sns.barplot(y=top_rated['Restaurant Name'], x=top_rated['Votes'], palette='coolwarm')
plt.xlabel('Votes')
plt.ylabel('Restaurant')
plt.title('Top 10 Most Rated Restaurants')
plt.show()
```

#  Analysis:
- The restaurant "Toit" has the highest number of votes, indicating strong customer engagement.  
- Highly rated restaurants tend to be well-known chains or popular local spots.

# 10. Bar Plot Top 10 Localities with the Most Restaurants

```python
plt.figure(figsize=(8,4))
df['Locality'].value_counts().head(10).plot(kind='bar', color='green')
plt.xlabel('Locality')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Localities with Most Restaurants')
plt.xticks(rotation=45)
plt.show()
```

#  Analysis:
- Connaught Place has the highest number of restaurants, making it a prime food hub. Other top localities also have a high density of restaurants, indicating competitive markets.
